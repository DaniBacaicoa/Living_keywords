````python
def build_prompt(topic, n_keywords):
    return f"""
    I will give you a topic. You need to provide the {n_keywords} most related keywords to this topic.
    For each keyword, assign a specificity score from 1 to 10 relative to the topic 
    (10 being extremely specific/synonymous, 1 being loosely related).
        
    Topic: {topic}
        
    Return the result purely as a JSON object with the following structure:
    {{
        "keywords": [
            {{"word": "keyword1", "score": 9}},
            {{"word": "keyword2", "score": 7}},
            ...
        ]
    }}
    Do not output any markdown formatting or extra text.
    """
````

````python
response = prompter.prompt(build_prompt("machine learning", 5))
````
